```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def func(x): 
    return np.cos(2*np.pi*x) + x**2 
def der(x):
    return -2*np.pi*np.sin(2*np.pi*x) + 2*x
```

```python
x = np.linspace(-2,2,1000)
y = func(x)
```

```python
plt.plot(x, y, x, der(x))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.legend(["f(x)", "f(x)'"])
plt.show()
```

```python
lr = 0.01
epochs = 1000 
#Khi to x l gi tr ngu nhin trong khong [-2,2]
localmin = np.random.choice(x,1)
# localmin = -2
print(localmin)
for i in range(epochs):
    localmin = localmin - lr*der(localmin)
print(localmin)
```

```python
plt.plot(x, y, x, der(x))
plt.plot(localmin, der(localmin), 'ro')
plt.plot(localmin, func(localmin), 'ro')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.legend(["f(x)", "f(x)'"])
plt.title(f"Emperical Local Minimum:{localmin[0]:.04}")
plt.show()
```
